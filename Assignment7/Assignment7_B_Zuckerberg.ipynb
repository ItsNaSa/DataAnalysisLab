```python
# Roll no: 33140 (Sahil Naphade)
# Batch: L9
# Assignment 7(B): WordCloud on Mark Zukerberg text fileimport pandas as pd
from wordcloud import wordcloud
import nltk
import numpy as np
from nltk import word_tokenize
import matplotlib.pyplot as pplt
from nltk.corpus import stopwords
# nltk.download('stopwords')
```

```python
data = open("G:/College/Sl-VI DataSets/TextMining/MarkZukerberg.txt","r").read()
data = data.lower()
print(data)
```

```python
stop = stopwords.words('english')
```

```python
# Function to remove numbers
def remove_num(text):
    text = ''.join([i for i in text if not i.isdigit()])
    return text

# Function to remove special characters and punctuations 
def remove_u(text):
    text = text.replace('_','')
    text = text.replace('?','')
    text = text.replace('•','')
    text = text.replace("@",'')
    text = text.replace('▯','')
    text = text.replace("'",'')
    text = text.replace('zkrbrg','')
    text = text.replace('[','')
    text = text.replace(']','')
    text = text.replace(",","")
    return text
```

```python
#executing functions to clean data
data = np.vectorize(remove_u)(data)
print(data)
```

```python
data = np.vectorize(remove_num)(data)
print(data)
```

```python
#removing extra spaces
def remove_extra_space(text):
    word_list = text.split()
    text = ' '.join(word_list)
    return text

#executing function
data = np.vectorize(remove_extra_space)(data)

# Convert data to list
data = data.tolist()
```

```python
def create_word_cloud(string):
    cloud = wordcloud.WordCloud(width=800,height=800,background_color = "white", stopwords = stop, max_words = 20).generate(string)
    pplt.figure(figsize = (10, 10), facecolor = None)
    pplt.axis('off')
    pplt.imshow(cloud)
```

```python
create_word_cloud(data)
```
